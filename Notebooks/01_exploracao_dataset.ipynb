# üê¶ Notebook 01 ‚Äì Explora√ß√£o do Dataset Xeno-Canto
Este notebook analisa o dataset utilizado no projeto MelodIA, incluindo visualiza√ß√£o,
estat√≠sticas de distribui√ß√£o, amostras de √°udio e espectrogramas.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import librosa, librosa.display
import numpy as np

metadata = pd.read_csv('data/xenocanto_meta.csv')
metadata.head()
```

```
metadata['cnt_species'] = metadata['gen'] + ' ' + metadata['sp']
plt.figure(figsize=(12,6))
metadata['cnt_species'].value_counts().head(20).plot(kind='bar')
plt.title('Top 20 esp√©cies mais frequentes')
plt.show()
```

## üéß Visualiza√ß√£o de espectrograma de um √°udio de exemplo

```
audio_path = 'samples/ave1.wav'
y, sr = librosa.load(audio_path)
plt.figure(figsize=(12,5))
S = librosa.feature.melspectrogram(y=y, sr=sr)
librosa.display.specshow(librosa.power_to_db(S, ref=np.max), sr=sr, x_axis='time', y_axis='mel')
plt.colorbar()
plt.show()
```
