# üé∂ Notebook 02 ‚Äì Extra√ß√£o de Embeddings com BirdNET
Este notebook mostra como gerar embeddings bioac√∫sticos usando BirdNET.

```
from birdnetlib import Recording
from birdnetlib.analyzer import Analyzer
import numpy as np
import librosa
import os

analyzer = Analyzer()
```

```
def extract_embedding(audio_path):
    recording = Recording(analyzer=analyzer, file_path=audio_path, lat=0.0, lon=0.0, week=20)
    recording.analyze()
    embs = [p['embedding'] for p in recording.predictions if 'embedding' in p]
    return np.mean(embs, axis=0)

embedding = extract_embedding('samples/ave1.wav')
embedding[:10]
```

## Visualiza√ß√£o via PCA / t-SNE (caso tenha v√°rias embeddings)

```
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

emb_list = []
files = ['samples/ave1.wav','samples/ave2.wav','samples/ave3.wav']

for f in files:
    emb_list.append(extract_embedding(f))

pca = PCA(n_components=2)
X = pca.fit_transform(emb_list)

plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1])
plt.title('PCA das embeddings')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```
