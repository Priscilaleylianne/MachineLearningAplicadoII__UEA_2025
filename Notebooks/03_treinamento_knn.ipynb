# üß† Notebook 03 ‚Äì Treinamento do KNN com Embeddings BirdNET
Este notebook treina o classificador KNN utilizando embeddings gerados pelo BirdNET.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
import pickle
```

## Carregar embeddings pr√©-processados

```
df = pd.read_pickle('data/embeddings.pkl')
X = list(df['embedding'])
y = df['species']

label_encoder = LabelEncoder()
y_enc = label_encoder.fit_transform(y)
```

## Treinar modelo KNN

```
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X, y_enc)

print('Treinamento conclu√≠do!')
```

## Avalia√ß√£o do modelo

```
y_pred = knn.predict(X)
print(classification_report(y_enc, y_pred))
```

```
cm = confusion_matrix(y_enc, y_pred)
plt.figure(figsize=(10,8))
plt.imshow(cm, cmap='Blues')
plt.title('Matriz de Confus√£o')
plt.colorbar()
plt.show()
```

## Salvar modelos

```
with open('models/knn_birdnet.pkl', 'wb') as f:
    pickle.dump(knn, f)

with open('models/label_encoder.pkl', 'wb') as f:
    pickle.dump(label_encoder, f)

print('Modelos salvos!')
```
